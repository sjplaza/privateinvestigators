```python
# Dependencies
import pandas as pd
import pprint as pprint
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create file path 
hate_crime_csv = "hate_crime.csv"
```

```python
# read the csv

hate_crime_df = pd.read_csv(hate_crime_csv, low_memory=False)
hate_crime_df.head()
```

```python
# filter by year 2014-2018 (5Years)

filtered_hate_crime_df = hate_crime_df[(hate_crime_df["DATA_YEAR"] >= 2014) & (hate_crime_df["DATA_YEAR"] <= 2018)]
filtered_hate_crime_df.head()
```

```python
# Sort the states by their STATE_NAME
state_df = filtered_hate_crime_df.sort_values("STATE_NAME")
state_df
```

```python
# Interesting Columns of Info for Presentation
columns_df = state_df[['INCIDENT_DATE', 'DATA_YEAR', 'PUB_AGENCY_NAME', 'STATE_NAME', 'POPULATION_GROUP_DESC', 'OFFENSE_NAME', 'BIAS_DESC']]
columns_df.head()
```

```python
# Making sure there is data in every column. If there were NA's, drop them.
columns_df.count()
```

```python
#Determine the amount of States. If there are any to ignore, do so.

len(columns_df["STATE_NAME"].unique())
```

```python
# Hate Crime counts for each state for the years of 
state_count_df = columns_df["STATE_NAME"].value_counts()
state_count_df
```

```python
len(columns_df["PUB_AGENCY_NAME"].unique())
```

```python
###### This is where I need to update the cities by the population to get refined data for populations over 250,000.

cities_count_df = columns_df.loc[(columns_df['POPULATION_GROUP_DESC'] == "Cities from 250,000 thru 499,999") | (columns_df['POPULATION_GROUP_DESC'] == "Cities from 500,000 thru 999,999") | (columns_df['POPULATION_GROUP_DESC'] == "Cities 1,000,000 or over"), ['INCIDENT_DATE', 'DATA_YEAR', 'PUB_AGENCY_NAME', 'STATE_NAME', 'POPULATION_GROUP_DESC', 'OFFENSE_NAME', 'BIAS_DESC'] ]
cities_count_df
```

```python
# Counts of all cities within the 250,000 pop criteria
pop_cities_df = cities_count_df[('PUB_AGENCY_NAME')].value_counts()
pop_cities_df
```

```python
# Showing the top 20 states
# Display the data with a clean layout
most_hate_crime_df = pd.DataFrame(pop_cities_df)
most_hate_crime_df.head(20)
```

```python
# Showing the bottom 20 states
# Display the data with a clean layout 
least_hate_crime_df = pd.DataFrame(pop_cities_df)
least_hate_crime_df.tail(20)
```
